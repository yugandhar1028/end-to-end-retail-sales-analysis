```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import to_datetime
```

```python
df=pd.read_csv(r"C:\Users\yugan\OneDrive\Documents\Ecommers data1.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'],format='%d-%m-%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'],format='%d-%m-%Y')
```

```python
df['OrderMonth']= df['Order Date'].dt.month
df['OrderYear']= df['Order Date'].dt.year
```

```python
df.info()
```

```python
df.describe()
```

```python
profitmargin= df['Profit'] / df['Total Sales']
print(profitmargin)
```

```python
Monthly_sales= df.groupby(['OrderMonth'])['Total Sales'].sum()
print(Monthly_sales)
```

```python
top_products=df.groupby(['Product Name'])['Total Sales'].sum().sort_values(ascending=False).head(5)
print(top_products)
```

```python
Discount_vs_profit=df[['Discount','Profit']].corr()
print(Discount_vs_profit)
```

```python
Monthly_sales.plot(kind='line',color='red')
plt.xlabel('Month')
plt.ylabel('sales')
plt.title('month vs sales')
plt.show()
```

```python
Region_vs_sales= df.groupby(['Region'])['Total Sales'].sum()
Region_vs_sales.plot(kind='bar',color='black')
plt.xlabel('region')
plt.ylabel('Sales')
plt.title('Region Vs Sales')
plt.show()
```

```python
Categort_vs_sales= df.groupby(['Category'])['Total Sales'].sum()
Categort_vs_sales.plot(kind='bar',color='black')
plt.xlabel('category')
plt.ylabel('Sales')
plt.title('category Vs Sales')
plt.show()
```

```python
Region_vs_Profit= df.groupby(['Region'])['Profit'].sum()
Region_vs_Profit.plot(kind='bar',color='black')
plt.xlabel('region')
plt.ylabel('Profit')
plt.title('Region_vs_Profit')
plt.show()
```
